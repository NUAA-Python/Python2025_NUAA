```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots() #fig(ure) ax(es)
# is more concise than this:
#fig = plt.figure()
#ax = fig.add_subplot(111)

print(type(fig))
print(type(ax))

squares = [1, 4, 9, 16, 25]
ax.plot(squares)

plt.show()
```

<h1 style="font-family:'Microsoft YaHei'; font-size:24px; color:purple;">
设置坐标轴标题和调整线宽
</h1>

<ul style="font-family: 'Microsoft YaHei'; font-size: 20px; line-height: 1.6;">
  <li><code>linewidth</code> 参数控制 <code>plot()</code> 绘制的<span style="color:blue;">线条粗细</span>。</li>
  <li><code>set_title()</code> 方法设置<span style="color:blue;">图表标题</span>。</li>
  <li><code>fontsize</code> 参数控制图表中<span style="color:blue;">文字的大小</span>。</li>
  <li><code>set_xlabel()</code> 和 <code>set_ylabel()</code> 方法设置<span style="color:blue;">坐标轴标题</span>。</li>
  <li><code>tick_params()</code> 方法设置<span style="color:blue;">刻度线样式</span>，这里将两个坐标轴的刻度文字大小设为 14。</li>
</ul>

<div style="text-align:center;">
  
</div>

```python
squares = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()

ax.plot(squares, linewidth=3)
# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=16)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)
# Set size of tick labels.
ax.tick_params(labelsize=14)

plt.show()
```

<h1 style="font-family:'Microsoft YaHei'; font-size:24px; color:purple;">
继续完善绘图
</h1>

<p style="font-family: 'Microsoft YaHei', Arial; font-size: 20px; line-height: 1.6;">
当只向 <code>plot()</code> 提供一组数值时，Python 会假设第一个数据点对应的 x 值为 0，但在我们的例子中，第一个数据点对应的 x 值应为 1。<br>
可以通过同时向 <code>plot()</code> 提供用于计算平方的输入值和输出值，来覆盖这种默认行为。<br>
When you give plot( ) a single sequence of numbers, it assumes the first data point corresponds to an x-value of 0, but our first point corresponds to an x-value of 1. We can override the default behaviour by giving plot( ) both the input and output values used to calculate the squares.
</p>

```python
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()

ax.plot(input_values, squares, linewidth=3)
# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=16)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)
# Set size of tick labels.
ax.tick_params(labelsize=14)

plt.show()
```

<span style="font-family:'Microsoft YaHei'; font-size:20px; color:purple; ">
扩展 3：可将绘制部分放在函数内
</span>

```python
import math
import matplotlib.pyplot as plt

def drawPlot(ax, stop_angle=360):
    # Include stop_angle in data
    xs = range(0, stop_angle + 1, 40)
    ys = [math.sin(math.radians(x)) for x in xs]   
    ax.plot(xs, ys, c="red") 

    # Set chart title and label axes
    ax.set_title("sin(x) function", fontsize=14)
    ax.set_xlabel("Angle x (deg)", fontsize=14)
    ax.set_ylabel("Value of sin(x)", fontsize=14)

    # Set x-ticks every 90 degrees
    xticks = range(0, stop_angle + 1, 90)
    ax.set_xticks(xticks)
    ax.set_xticklabels(xticks)


_, ax = plt.subplots(figsize=(12, 4))
drawPlot(ax)
plt.show()
```

<h2 style="font-family:'Microsoft YaHei'; font-size:24px; color:purple;">
随机游走（散点图）
</h2>
<p style="font-family: 'Microsoft YaHei', Arial; font-size: 20px; line-height: 1.6;">
使用 Python 生成随机游走数据，然后使用 Matplotlib 创建该数据的可视化表示。<br>
随机游走是一条由一系列简单决策决定的路径，每个决策完全由随机因素决定。你可以把随机游走想象成一只迷路的蚂蚁，如果它每一步都朝随机方向移动时的路径。
<br>
随机游走在自然、物理学、生物学、化学和经济学中都有实际应用。例如，漂浮在水面上的花粉颗粒会在水面上移动，因为水分子不断地推动它。水滴中分子的运动是随机的，因此花粉颗粒在水面上描绘的路径就是一次随机游走。
</p>

```python
from random import choice

class RandomWalk:
    """A class to generate random walks."""

    def __init__(self, num_points=5000):
        """Initialize attributes of a walk."""
        self.num_points = num_points

        # All walks start at (0, 0).
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """Calculate all the points in the walk."""
        # Keep taking steps until the walk reaches the desired length.
        while len(self.x_values) < self.num_points:
            # Decide which direction to go, and how far to go.
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance

            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance

            # Reject moves that go nowhere.
            if x_step == 0 and y_step == 0:
                continue

            # Calculate the new position.
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step

            self.x_values.append(x)
            self.y_values.append(y)

# Make a random walk.
rw = RandomWalk(100) #1_000
rw.fill_walk()

# Plot the points in the walk.
fig, ax = plt.subplots(figsize=(15, 9)) #dimensions of the plotting window in inches
point_numbers = range(rw.num_points)
ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues,
                edgecolors='none', s=15)
ax.set_aspect('equal') #坐标轴的纵横比为 1:1

# Emphasize the first and last points.
ax.scatter(0, 0, c='green', edgecolors='none', s=100)
ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none',
                s=100)

plt.show()
```

<h2 style="font-family:'Microsoft YaHei'; font-size:22px; color:purple;">
fill_between 函数
</h2>
<span style="font-family:'Microsoft YaHei'; font-size:20px; color:black;">
The Axes.fill_between() function in axes module of matplotlib library is used to <span style="color:blue;">fill the area between two horizontal curves</span>.<br>
<code>Axes.fill_between(self, x, y1, y2=0, where=None, interpolate=False, step=None, *, data=None, **kwargs)</code><br>
x: This parameter contains the horizontal coordinates of the data points which are used to define the curves.<br>
y1: This parameter contains the y coordinates of the data points which are used to define the first curves<br>
y2: This parameter contains the y coordinates of the data points which are used to define the second curves. <br>
It is an optional with default value of 0.
</span>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 1000)
y1 = np.sqrt(1-(abs(x)-1)**2)
y2 = -3*np.sqrt(1-(abs(x)/2)**0.5)

fig, ax = plt.subplots()

ax.fill_between(x, y1, color='red')
ax.fill_between(x, y2, color='red')
ax.set_aspect('equal')
#ax.axis('off')  # optional: hide axes

plt.show()
```

<span style="font-family:'Microsoft YaHei'; font-size:22px; color:purple;">
示例 1：棋盘
</span>

```python
import matplotlib.pyplot as plt
import numpy as np

size = 8
checkerboard = np.zeros((size, size))

# Fill with checker pattern
for i in range(size):
    for j in range(size):
        if (i + j) % 2 == 0:
            checkerboard[i, j] = 1.0  # White squares

# Plot
fig, ax = plt.subplots(figsize=(5, 5))

# Use 'nearest' interpolation to keep the squares sharp
ax.imshow(checkerboard, cmap='gray', interpolation='nearest')

ax.set_title("Checkerboard Pattern", fontsize=14)
ax.set_xticks([]) #an empty list
ax.set_yticks([])

plt.show()
```

<span style="font-family:'Microsoft YaHei'; font-size:22px; color:purple;">

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)
Z = np.sin(8*theta + 3*R) * np.exp(-0.1*R**2)

fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(Z, cmap='cool', origin='lower')
ax.set_title("Swirling Vortex (Pseudo-Color)", fontsize=14)
ax.set_xticks([]); ax.set_yticks([])
plt.colorbar(im, ax=ax, fraction=0.045, pad=0.04)
plt.show()
```
